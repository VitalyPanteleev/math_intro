```python
def vector_length(inp_vector):
    res = 0
    for el in inp_vector: res += el**2
    return res**0.5

some_vector = [3, 4]

vector_length(some_vector)
```

```python
def polar_to_rect(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x, y

print (polar_to_rect(3, np.pi / 4))
```
